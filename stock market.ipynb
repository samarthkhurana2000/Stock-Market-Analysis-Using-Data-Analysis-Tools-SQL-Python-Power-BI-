```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/gm_100_samarth_python.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['no_analyst_rec'].unique()
```

```python
# This code is to fill the null values
df['no_analyst_rec'] = df['no_analyst_rec'].fillna(df['no_analyst_rec'].mode().values[0])
df = df[~ df['no_analyst_rec'].isin(['0', ' ',])]
```

```python
df.info()
```

```python

df['sub_sector'].unique()
```

```python
df['sub_sector'] = df['sub_sector'].fillna(df['sub_sector'].mode().values[0])
df = df[~ df['sub_sector'].isin(['0', ' ',])]
df['sub_sector'].unique()
```

```python
#Below code is for the calculation of ratios
df['pat'] = df['PBT'] - df['Taxes_others']
df['outstanding_shares'] = df['market_cap'] - df['pat']
df['eps'] = df['pat'] / df['outstanding_shares']
df['percentage_hold_by_promoter'] = (df['promoter_holding'] / df['close_price']) * 100
df['analyst_buy_ratio'] = (df['no_analyst_rec'] / df['percentage_buy']) * 100
df['roe'] = df['pat'] / (df['total_revenue'] - df['pat']) * 100
df['profit_margin'] = (df['pat'] / df['total_revenue']) * 100
df['ps_ratio'] = df['market_cap'] / df['total_revenue']
df['retail_holding_change'] = -(df['dii_holding'] + df['fii_holding'] + df['mf_holding'] + df['promoter_holding'])
```

```python
#Below code is to round off the values
df['pat'] = df['pat'].round(2)
df['eps'] = df['eps'].round(2)
df['analyst_buy_ratio'] = df['analyst_buy_ratio'].round(2)
df['roe'] = df['roe'].round(2)
df['percentage_hold_by_promoter'] = df['percentage_hold_by_promoter'].round(2)
df['profit_margin'] = df['profit_margin'].round(2)
```

```python
#Below code is to convert this file into csv to import it into the POWER BI Desktop for visualization
df.to_csv('/content/gm_100_samarth.csv', index=False)
```
